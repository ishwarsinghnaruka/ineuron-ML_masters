1.
Create the below pattern using nested for loop in python

```python
##Method_1 - without nested for loop
for i in range(1,10):
        if i<6:
            print('* '*i)
        else:
            print('* '*(10-i))
```

```python
## Method_2_using nested for loop
for i in range(5):
    for j in range(i):
        print ('* ', end="")
    print('')
for i in range(5,0,-1):
    for j in range(i):
        print('* ', end="")
    print('')
```
